```python
import pandas as pd
import pickle
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score, confusion_matrix, ConfusionMatrixDisplay
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestClassifier
from sklearn.naive_bayes import MultinomialNB
from sklearn.linear_model import SGDClassifier
```

```python
# Veri setini yükleyin
df = pd.read_csv("processed_nachecked_data.csv")

# Haber gövdesi ve sınıfı sütunlarını alın
haberler = df["Haber Gövdesi"]
siniflar = df["Sınıf"]
```

```python
# TF-IDF vektörleştiriciyi oluşturun
tfidf_vectorizer = TfidfVectorizer(max_features=1000)  # En önemli 1000 kelime

# Haber gövdelerini TF-IDF matrisine dönüştürün
tfidf_matrix = tfidf_vectorizer.fit_transform(haberler)
```

```python
# Veriyi eğitim ve test setlerine ayırın
X_train, X_test, y_train, y_test = train_test_split(
    tfidf_matrix, siniflar, test_size=0.2, random_state=42
)

# Logistic Regression modelini eğitin
classifier = LogisticRegression(max_iter=1000)
classifier.fit(X_train, y_train)
```

```python
# Test seti üzerinde tahmin yapın
y_pred = classifier.predict(X_test)

# Modelin performansını değerlendirin
print(classification_report(y_test, y_pred))

cm = confusion_matrix(y_test, y_pred, labels=classifier.classes_)

# Confusion matrix'i görselleştirin
fig, ax = plt.subplots(figsize=(17, 14))  # Burada figsize parametresini daha büyük yapıyoruz
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=classifier.classes_)
disp.plot(cmap=plt.cm.Blues, ax=ax)
plt.show()
```

```python
with open("logistic_regression_model.pkl", "wb") as model_file:
    pickle.dump(classifier, model_file)

with open("tfidf_vectorizer.pkl", "wb") as vectorizer_file:
    pickle.dump(tfidf_vectorizer, vectorizer_file)
```

```python
X_train, X_test, y_train, y_test = train_test_split(tfidf_matrix, siniflar, test_size=0.2, random_state=42)

# Naive Bayes modelini oluşturun ve eğitin
nb_model = MultinomialNB()
nb_model.fit(X_train, y_train)
```

```python
y_pred = nb_model.predict(X_test)

# Modelin performansını değerlendirin
print(classification_report(y_test, y_pred))

cm = confusion_matrix(y_test, y_pred, labels=nb_model.classes_)

# Confusion matrix'i görselleştirin
fig, ax = plt.subplots(figsize=(17, 14))  # Burada figsize parametresini daha büyük yapıyoruz
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=nb_model.classes_)
disp.plot(cmap=plt.cm.Blues, ax=ax)
plt.show()
```

```python
X_train, X_test, y_train, y_test = train_test_split(tfidf_matrix, siniflar, test_size=0.2, random_state=42)

# SGD Classifier modelini oluşturun ve eğitin
sgd_model = SGDClassifier(loss='log_loss', max_iter=1000, tol=1e-3, random_state=42)
sgd_model.fit(X_train, y_train)
```

```python
# Test verisi ile tahmin yapın
y_pred = sgd_model.predict(X_test)

print(classification_report(y_test, y_pred))

cm = confusion_matrix(y_test, y_pred, labels=sgd_model.classes_)

# Confusion matrix'i görselleştirin
fig, ax = plt.subplots(figsize=(17, 14))  # Burada figsize parametresini daha büyük yapıyoruz
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=sgd_model.classes_)
disp.plot(cmap=plt.cm.Blues, ax=ax)
plt.show()
```
